## Feature scaling

```python
scaler = StandardScaler()

df['scaled_amount'] = scaler.fit_transform(df['Amount'].values.reshape(-1,1))

df = df.drop(['Amount'], axis=1)
```

```python
scaler
```

## Feature And Labels

```python
x=df.drop('Class',axis=1)
y=df['Class']
```

```python
x
```

```python
y
```

## Train Test Split data

```python
non_nan_indices = y.dropna().index
x_clean = x.loc[non_nan_indices]
y_clean = y.loc[non_nan_indices]
x_train,x_test,y_train,y_test=train_test_split(x_clean,y_clean,test_size=0.2,random_state=42 ,stratify=y_clean)
```

## Handle Imbalanced Data Using SMOTE

```python
smote = SMOTE(random_state=42)
x_train_resampled, y_train_resampled = smote.fit_resample(x_train, y_train)
```

```python
smote
```

```python
smote = SMOTE(random_state=42)

x_train_smote, y_train_smote = smote.fit_resample(x_train, y_train)
```

```python
smote
```

## Train Model

```python
model= LogisticRegression()
model.fit(x_train_smote,y_train_smote)
```

## Random Forest

```python
model= RandomForestClassifier()
model.fit(x_train_smote,y_train_smote)
```

## Prediction

```python
y_pred=model.predict(x_test)
```

## Evaluate model

```python
confusion_matrix(y_test,y_pred)
classification_matrix(y_test,y_pred)
accuracy_score(y_test,y_pred)
precision_score(y_test,y_pred)
recall_score(y_test,y_pred)
```

**ROC**-**AUC**

```python
y_prob=model.predict_proba(x_test)[:,1]
roc_auc_score(y_test,y_prob)
```

## Visualize Confusion Matrix

```python
Confusion_Matrics=confusion_matrix(y_test,y_pred)
sns.heatmap(Confusion_Matrics,annot=True,fmt='d')
plt.show()
```
