```python
# Feature & labels
```

```python
features = []
labels = []

emotion_dict = {
    '01': 'neutral',
    '02': 'calm',
    '03': 'happy',
    '04': 'sad',
    '05': 'angry',
    '06': 'fearful',
    '07': 'disgust',
    '08': 'surprised'
}

for file_path in wav_files:

    file_name = file_path.split('/')[-1]

    emotion_code = file_name.split('-')[2]

    emotion = emotion_dict[emotion_code]

    audio, sr = librosa.load(file_path, duration=3)

    mfcc = librosa.feature.mfcc(
        y=audio,
        sr=sr,
        n_mfcc=40
    )

    mfcc_features = np.mean(mfcc.T, axis=0)

    features.append(mfcc_features)
    labels.append(emotion)
```

```python
print("Total Features:", len(features))
print("Total Labels:", len(labels))
print(labels[:5])
```

```python
# convert to numpy array
X = np.array(features)
y = np.array(labels)

print(X.shape)
print(y.shape)
```

```python
print(X.shape)
print(y.shape)
```

```python
# label Encoding 
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()

y_encoded = encoder.fit_transform(y)

print(y[:5])
print(y_encoded[:5])
```

```python
from tensorflow.keras.utils import to_categorical

y_categorical = to_categorical(y_encoded)

print(y_categorical.shape)
```

## Train-Test Split

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y_categorical,
    test_size=0.2,
    random_state=42
)

print(X_train.shape)
print(X_test.shape)
```

```python
X_train = X_train.reshape(X_train.shape[0], 40, 1)
X_test = X_test.reshape(X_test.shape[0], 40, 1)

print(X_train.shape)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
```

## Bi-GRU architecture

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Bidirectional, GRU, Dense, Dropout

model = Sequential()

# Bi-GRU Layer
model.add(Bidirectional(GRU(64, return_sequences=False),
                        input_shape=(40,1)))

# Dropout
model.add(Dropout(0.3))

# Hidden Layer
model.add(Dense(32, activation='relu'))

# Output Layer
model.add(Dense(8, activation='softmax'))

model.summary()
```

```python
Dropout(0.3)
```

## Compile Model

```python
model.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)
```

```python
history = model.fit(
    X_train,
    y_train,
    epochs=30,
    batch_size=32,
    validation_data=(X_test, y_test)
)
```

```python
loss, accuracy = model.evaluate(X_test, y_test)

print("Accuracy:", accuracy)
```
